```python
import os
from typing import TypedDict
from google.oauth2 import service_account
from langchain_community.tools.tavily_search import TavilySearchResults
from langchain_google_vertexai import ChatVertexAI
from IPython.display import Image, display, Markdown
from langgraph.graph import StateGraph, END

# ‡∫ï‡∫±‡ªâ‡∫á‡∫Ñ‡ªà‡∫≤ API Keys
TAVILY_API_KEY = os.getenv("TAVILY_API_KEY")  # ‡∫õ‡ªà‡∫Ω‡∫ô‡ªÄ‡∫õ‡∫±‡∫ô API Key ‡∫à‡∫¥‡∫á‡∫Ç‡∫≠‡∫á‡∫ó‡ªà‡∫≤‡∫ô

# ‡∫ï‡∫±‡ªâ‡∫á‡∫Ñ‡ªà‡∫≤ Google Vertex AI
JSON_KEY_PATH = r"C:\Users\Kenshin\Desktop\GoogleI-O\key.json"
PROJECT_ID = os.getenv("GOOGLE_CLOUD_PROJECT")
GOOGLE_LOCATION = os.getenv("GOOGLE_LOCATION")

credentials = service_account.Credentials.from_service_account_file(
    JSON_KEY_PATH,
    scopes=["https://www.googleapis.com/auth/cloud-platform"]
)

# ‡ªÄ‡∫•‡∫∑‡∫≠‡∫Å‡ªÇ‡∫°‡ªÄ‡∫î‡∫• Gemini
model = ChatVertexAI(
    model_name="gemini-2.5-flash-lite",
    project=PROJECT_ID,
    location=GOOGLE_LOCATION,
    credentials=credentials,
    temperature=1.0
)

class AgentState(TypedDict):
    question: str
    search_results: list
    content: str
    user_answers: list
    current_question_num: int
    total_questions: int
    feedback: str
    is_correct: bool
    all_correct: bool

# ‡∫™‡ªâ‡∫≤‡∫á Node ‡∫Ñ‡∫ª‡ªâ‡∫ô‡∫´‡∫≤‡∫à‡∫≤‡∫Å‡∫≠‡∫¥‡∫ô‡ªÄ‡∫ï‡∫µ‡ªÄ‡∫ô‡∫±‡∫î (5 ‡ªÅ‡∫´‡∫º‡ªà‡∫á)
def web_search(state: AgentState):
    try:
        # ‡∫™‡ªâ‡∫≤‡∫á Tavily search tool
        search = TavilySearchResults(api_key=TAVILY_API_KEY)
        
        # ‡∫Ñ‡∫ª‡ªâ‡∫ô‡∫´‡∫≤‡∫Ç‡ªç‡ªâ‡∫°‡∫π‡∫ô‡∫ó‡∫µ‡ªà‡∫Å‡ªà‡∫Ω‡∫ß‡∫Ç‡ªâ‡∫≠‡∫á
        results = search.invoke({
            "query": f"IELTS reading topic {state['question']} site:cambridge.org OR site:ielts.org",
            "max_results": 5  # ‡∫à‡∫≥‡∫Å‡∫±‡∫î‡∫û‡∫Ω‡∫á 5 ‡∫ú‡∫ª‡∫ô‡∫Å‡∫≤‡∫ô‡∫Ñ‡∫ª‡ªâ‡∫ô‡∫´‡∫≤
        })
        
        state['search_results'] = results
        return state
    
    except Exception as e:
        display(Markdown(f"**Error:** <span style='color:red'>‡ªÄ‡∫Å‡∫µ‡∫î‡∫Ç‡ªç‡ªâ‡∫ú‡∫¥‡∫î‡∫û‡∫≤‡∫î‡ªÉ‡∫ô‡∫Å‡∫≤‡∫ô‡∫Ñ‡∫ª‡ªâ‡∫ô‡∫´‡∫≤: {str(e)}</span>"))
        state['search_results'] = []
        return state

# ‡∫™‡ªâ‡∫≤‡∫á Node ‡∫™‡ªâ‡∫≤‡∫á‡∫ö‡∫ª‡∫î‡∫Ñ‡∫ß‡∫≤‡∫°
def create_reading_content(state: AgentState):
    try:
        if not state['search_results']:
            raise ValueError("‡∫ö‡ªç‡ªà‡∫°‡∫µ‡∫ú‡∫ª‡∫ô‡∫Å‡∫≤‡∫ô‡∫Ñ‡∫ª‡ªâ‡∫ô‡∫´‡∫≤")
            
        # ‡∫™‡ªâ‡∫≤‡∫á‡∫Ñ‡∫≥‡∫≠‡∫∞‡∫ó‡∫¥‡∫ö‡∫≤‡∫ç‡∫à‡∫≤‡∫Å‡∫ú‡∫ª‡∫ô‡∫Å‡∫≤‡∫ô‡∫Ñ‡∫ª‡ªâ‡∫ô‡∫´‡∫≤
        sources_text = "\n\n".join([
            f"‡ªÅ‡∫´‡∫º‡ªà‡∫á‡∫ó‡∫µ‡ªà {i+1}:\nURL: {result['url']}\n‡ªÄ‡∫ô‡∫∑‡ªâ‡∫≠‡∫´‡∫≤: {result['content'][:200]}..." 
            for i, result in enumerate(state['search_results'])
        ])
        
        # ‡ªÉ‡∫ä‡ªâ LLM ‡∫™‡ªâ‡∫≤‡∫á‡∫ö‡∫ª‡∫î‡∫Ñ‡∫ß‡∫≤‡∫° Reading IELTS
        prompt = f"""
        ‡∫à‡∫≤‡∫Å‡∫ú‡∫ª‡∫ô‡∫Å‡∫≤‡∫ô‡∫Ñ‡∫ª‡ªâ‡∫ô‡∫´‡∫≤ 5 ‡ªÅ‡∫´‡∫º‡ªà‡∫á‡∫ï‡ªç‡ªà‡ªÑ‡∫õ‡∫ô‡∫µ‡ªâ ‡∫Å‡ªà‡∫Ω‡∫ß‡∫Å‡∫±‡∫ö‡∫´‡∫ª‡∫ß‡∫Ç‡ªç‡ªâ '{state['question']}', 
        ‡∫Å‡∫∞‡∫•‡∫∏‡∫ô‡∫≤‡∫™‡ªâ‡∫≤‡∫á‡∫ö‡∫ª‡∫î‡∫Ñ‡∫ß‡∫≤‡∫° Reading IELTS ‡∫ó‡∫µ‡ªà‡ªÄ‡ªù‡∫≤‡∫∞‡∫™‡∫ª‡∫°:

        {sources_text}

        ‡∫Ç‡ªç‡ªâ‡∫Å‡∫≥‡∫ô‡∫ª‡∫î:
        1. ‡∫Ç‡∫Ω‡∫ô‡ªÄ‡∫õ‡∫±‡∫ô‡∫û‡∫≤‡∫™‡∫≤‡∫≠‡∫±‡∫á‡∫Å‡∫¥‡∫î ‡∫•‡∫∞‡∫î‡∫±‡∫ö IELTS Academic
        2. ‡∫ç‡∫≤‡∫ß‡∫õ‡∫∞‡∫°‡∫≤‡∫ô 300-400 ‡∫Ñ‡∫≥
        3. ‡∫™‡ªâ‡∫≤‡∫á‡∫Ñ‡∫≥‡∫ñ‡∫≤‡∫°‡∫û‡∫Ω‡∫á 2 ‡∫õ‡∫∞‡ªÄ‡∫û‡∫î‡ªÄ‡∫ó‡∫ª‡ªà‡∫≤‡∫ô‡∫±‡ªâ‡∫ô:
           - **Sentence Completion** (3 ‡∫Ç‡ªç‡ªâ): ‡∫™‡ªâ‡∫≤‡∫á‡∫õ‡∫∞‡ªÇ‡∫´‡∫ç‡∫Å‡∫ó‡∫µ‡ªà‡∫ö‡ªç‡ªà‡∫™‡∫ª‡∫°‡∫ö‡∫π‡∫ô ‡ªÅ‡∫•‡∫∞ ‡ªÉ‡∫´‡ªâ‡∫ú‡∫π‡ªâ‡ªÉ‡∫ä‡ªâ‡∫ï‡∫∑‡ªà‡∫°‡∫Ñ‡∫≥‡∫à‡∫≤‡∫Å‡∫ö‡∫ª‡∫î‡∫Ñ‡∫ß‡∫≤‡∫° (‡∫ö‡ªç‡ªà‡ªÄ‡∫Å‡∫µ‡∫ô 3 ‡∫Ñ‡∫≥)
           - **True/False/Not Given** (2 ‡∫Ç‡ªç‡ªâ): ‡∫™‡ªâ‡∫≤‡∫á‡∫õ‡∫∞‡ªÇ‡∫´‡∫ç‡∫Å‡ªÄ‡∫û‡∫∑‡ªà‡∫≠‡ªÉ‡∫´‡ªâ‡∫ú‡∫π‡ªâ‡ªÉ‡∫ä‡ªâ‡∫Å‡∫ß‡∫î‡∫™‡∫≠‡∫ö‡∫ß‡ªà‡∫≤ True, False, ‡∫´‡∫º‡∫∑ Not Given
        4. ‡ªÄ‡∫ô‡∫±‡ªâ‡∫ô‡∫Å‡∫≤‡∫ô‡∫ô‡∫≥‡ªÉ‡∫ä‡ªâ‡∫Ç‡ªç‡ªâ‡∫°‡∫π‡∫ô‡∫à‡∫≤‡∫Å‡ªÅ‡∫´‡∫º‡ªà‡∫á‡∫ó‡∫µ‡ªà‡∫ô‡ªç‡∫≤‡ªÄ‡∫ä‡∫∑‡ªà‡∫≠‡∫ñ‡∫∑‡ªÑ‡∫î‡ªâ
        
        **‡∫™‡∫≥‡∫Ñ‡∫±‡∫ô: ‡ªÅ‡∫ç‡∫Å‡∫ú‡∫ª‡∫ô‡ªÑ‡∫ô‡ªÉ‡∫´‡ªâ‡∫ä‡∫±‡∫î‡ªÄ‡∫à‡∫ô**
        
        ‡∫à‡∫±‡∫î‡∫Æ‡∫π‡∫ö‡ªÅ‡∫ö‡∫ö‡∫î‡∫±‡ªà‡∫á‡∫ô‡∫µ‡ªâ:
        
        === READING PASSAGE ===
        [‡∫ö‡∫ª‡∫î‡∫Ñ‡∫ß‡∫≤‡∫° 300-400 ‡∫Ñ‡∫≥]
        
        === QUESTIONS ===
        **Sentence Completion (Choose NO MORE THAN THREE WORDS from the passage):**
        1. The research shows that _______ is the main cause of...
        2. Scientists discovered that _______ can improve...
        
        **True/False/Not Given:**
        3. The study was conducted over a period of five years.
        4. All participants showed positive results.
        5. The research was funded by government agencies.
        
        === ANSWER_KEY_INTERNAL ===
        1. [exact words from passage]
        2. [exact words from passage]  
        3. True/False/Not Given
        4. True/False/Not Given
        5. True/False/Not Given
        """
        
        response = model.invoke(prompt)
        full_content = response.content
        
        # ‡ªÅ‡∫ç‡∫Å‡ªÄ‡∫ô‡∫∑‡ªâ‡∫≠‡∫´‡∫≤‡∫≠‡∫≠‡∫Å‡∫à‡∫≤‡∫Å Answer Key
        if "=== ANSWER_KEY_INTERNAL ===" in full_content:
            content_parts = full_content.split("=== ANSWER_KEY_INTERNAL ===")
            # ‡ªÄ‡∫≠‡∫ª‡∫≤‡ªÄ‡∫â‡∫û‡∫≤‡∫∞‡∫™‡ªà‡∫ß‡∫ô‡∫ó‡∫µ‡ªà‡∫à‡∫∞‡∫™‡∫∞‡ªÅ‡∫î‡∫á‡ªÉ‡∫´‡ªâ‡∫ú‡∫π‡ªâ‡ªÉ‡∫ä‡ªâ
            state['content'] = content_parts[0].replace("=== READING PASSAGE ===", "**Reading Passage:**").replace("=== QUESTIONS ===", "**Questions:**").strip()
            # ‡ªÄ‡∫Å‡∫±‡∫ö Answer Key ‡ªÑ‡∫ß‡ªâ‡∫û‡∫≤‡∫ç‡ªÉ‡∫ô
            state['answer_key'] = content_parts[1].strip() if len(content_parts) > 1 else ""
        else:
            state['content'] = full_content
            state['answer_key'] = ""
        return state
    
    except Exception as e:
        display(Markdown(f"**Error:** <span style='color:red'>‡ªÄ‡∫Å‡∫µ‡∫î‡∫Ç‡ªç‡ªâ‡∫ú‡∫¥‡∫î‡∫û‡∫≤‡∫î‡ªÉ‡∫ô‡∫Å‡∫≤‡∫ô‡∫™‡ªâ‡∫≤‡∫á‡∫ö‡∫ª‡∫î‡∫Ñ‡∫ß‡∫≤‡∫°: {str(e)}</span>"))
        state['content'] = ""
        return state

# ‡∫™‡ªâ‡∫≤‡∫á Node ‡∫Å‡∫ß‡∫î‡∫™‡∫≠‡∫ö‡∫Ñ‡∫≥‡∫ï‡∫≠‡∫ö
def check_answer(state: AgentState):
    try:
        # ‡∫Å‡∫ß‡∫î‡∫™‡∫≠‡∫ö‡∫Ñ‡∫≥‡∫ï‡∫≠‡∫ö‡∫ó‡∫±‡∫á‡ªù‡∫ª‡∫î
        if 'user_answers' in state and len(state['user_answers']) >= 5:
            prompt = f"""
            ‡∫ö‡∫ª‡∫î‡∫Ñ‡∫ß‡∫≤‡∫° Reading:
            {state['content']}
            
            ‡∫Ñ‡∫≥‡∫ï‡∫≠‡∫ö‡∫Ç‡∫≠‡∫á‡∫ô‡∫±‡∫Å‡∫Æ‡∫Ω‡∫ô (‡∫ï‡∫≤‡∫°‡∫•‡∫≥‡∫î‡∫±‡∫ö 1-5):
            {', '.join([f"{i+1}. {ans}" for i, ans in enumerate(state['user_answers'])])}
            
            ‡∫Å‡∫∞‡∫•‡∫∏‡∫ô‡∫≤‡∫Å‡∫ß‡∫î‡∫™‡∫≠‡∫ö‡∫Ñ‡∫≥‡∫ï‡∫≠‡∫ö‡∫ó‡∫±‡∫á‡ªù‡∫ª‡∫î‡∫ï‡∫≤‡∫°‡ªÄ‡∫á‡∫∑‡ªà‡∫≠‡∫ô‡ªÑ‡∫Ç IELTS Reading:
            
            **‡∫™‡∫≥‡∫´‡∫º‡∫±‡∫ö Sentence Completion:**
            - ‡∫Å‡∫ß‡∫î‡∫™‡∫≠‡∫ö‡∫ß‡ªà‡∫≤‡∫Ñ‡∫≥‡∫ï‡∫≠‡∫ö‡∫°‡∫µ‡∫ö‡ªç‡ªà‡ªÄ‡∫Å‡∫µ‡∫ô 3 ‡∫Ñ‡∫≥
            - ‡∫Ñ‡∫≥‡∫ï‡∫≠‡∫ö‡∫ï‡ªâ‡∫≠‡∫á‡∫°‡∫≤‡∫à‡∫≤‡∫Å‡∫ö‡∫ª‡∫î‡∫Ñ‡∫ß‡∫≤‡∫°‡ªÇ‡∫î‡∫ç‡∫Å‡∫ª‡∫á
            - ‡∫ï‡ªâ‡∫≠‡∫á‡∫ñ‡∫∑‡∫Å‡∫ï‡ªâ‡∫≠‡∫á‡∫ó‡∫≤‡∫á‡∫î‡ªâ‡∫≤‡∫ô‡ªÑ‡∫ß‡∫ç‡∫≤‡∫Å‡∫≠‡∫ô
            
            **‡∫™‡∫≥‡∫´‡∫º‡∫±‡∫ö True/False/Not Given:**
            - True: ‡∫ñ‡ªâ‡∫≤‡∫Ç‡ªç‡ªâ‡∫Ñ‡∫ß‡∫≤‡∫°‡∫Å‡∫ª‡∫á‡∫Å‡∫±‡∫ö‡∫ö‡∫ª‡∫î‡∫Ñ‡∫ß‡∫≤‡∫°
            - False: ‡∫ñ‡ªâ‡∫≤‡∫Ç‡ªç‡ªâ‡∫Ñ‡∫ß‡∫≤‡∫°‡∫Å‡∫ª‡∫á‡∫Å‡∫±‡∫ô‡∫Ç‡ªâ‡∫≤‡∫°‡∫Å‡∫±‡∫ö‡∫ö‡∫ª‡∫î‡∫Ñ‡∫ß‡∫≤‡∫°  
            - Not Given: ‡∫ñ‡ªâ‡∫≤‡∫ö‡ªç‡ªà‡∫°‡∫µ‡∫Ç‡ªç‡ªâ‡∫°‡∫π‡∫ô‡ªÉ‡∫ô‡∫ö‡∫ª‡∫î‡∫Ñ‡∫ß‡∫≤‡∫°
            
            ‡∫Å‡∫∞‡∫•‡∫∏‡∫ô‡∫≤‡ªÉ‡∫´‡ªâ‡∫ú‡∫ª‡∫ô‡∫Å‡∫≤‡∫ô‡∫Å‡∫ß‡∫î‡∫™‡∫≠‡∫ö‡ªÅ‡∫ï‡ªà‡∫•‡∫∞‡∫Ç‡ªç‡ªâ:
            1. ‡∫ñ‡∫∑‡∫Å/‡∫ú‡∫¥‡∫î ‡ªÅ‡∫•‡∫∞ ‡ªÄ‡∫´‡∫î‡∫ú‡∫ª‡∫ô
            2. ‡∫ñ‡∫∑‡∫Å/‡∫ú‡∫¥‡∫î ‡ªÅ‡∫•‡∫∞ ‡ªÄ‡∫´‡∫î‡∫ú‡∫ª‡∫ô
            3. ‡∫ñ‡∫∑‡∫Å/‡∫ú‡∫¥‡∫î ‡ªÅ‡∫•‡∫∞ ‡ªÄ‡∫´‡∫î‡∫ú‡∫ª‡∫ô
            4. ‡∫ñ‡∫∑‡∫Å/‡∫ú‡∫¥‡∫î ‡ªÅ‡∫•‡∫∞ ‡ªÄ‡∫´‡∫î‡∫ú‡∫ª‡∫ô
            5. ‡∫ñ‡∫∑‡∫Å/‡∫ú‡∫¥‡∫î ‡ªÅ‡∫•‡∫∞ ‡ªÄ‡∫´‡∫î‡∫ú‡∫ª‡∫ô
            
            ‡∫ñ‡ªâ‡∫≤‡∫°‡∫µ‡∫Ç‡ªç‡ªâ‡ªÉ‡∫î‡∫ú‡∫¥‡∫î, ‡ªÉ‡∫´‡ªâ Feedback ‡ªÅ‡∫•‡∫∞ Highlight ‡∫™‡ªà‡∫ß‡∫ô‡∫ó‡∫µ‡ªà‡∫Å‡ªà‡∫Ω‡∫ß‡∫Ç‡ªâ‡∫≠‡∫á.
            ‡∫ñ‡ªâ‡∫≤‡∫ó‡∫∏‡∫Å‡∫Ç‡ªç‡ªâ‡∫ñ‡∫∑‡∫Å‡∫ï‡ªâ‡∫≠‡∫á, ‡ªÉ‡∫´‡ªâ‡∫ï‡∫≠‡∫ö‡∫ß‡ªà‡∫≤ "ALL CORRECT" ‡∫¢‡ªà‡∫≤‡∫á‡∫ä‡∫±‡∫î‡ªÄ‡∫à‡∫ô.
            """
            
            response = model.invoke(prompt)
            feedback = response.content
            
            # ‡∫Å‡∫ß‡∫î‡∫™‡∫≠‡∫ö‡∫ß‡ªà‡∫≤‡∫ó‡∫∏‡∫Å‡∫Ç‡ªç‡ªâ‡∫ñ‡∫∑‡∫Å‡∫ï‡ªâ‡∫≠‡∫á‡∫´‡∫º‡∫∑‡∫ö‡ªç‡ªà
            if "ALL CORRECT" in feedback.upper():
                state['all_correct'] = True
                state['is_correct'] = True
            else:
                state['all_correct'] = False
                state['is_correct'] = False
                state['feedback'] = feedback
        else:
            # ‡∫ç‡∫±‡∫á‡∫ï‡∫≠‡∫ö‡∫ö‡ªç‡ªà‡∫Ñ‡∫ª‡∫ö
            state['is_correct'] = False
            state['all_correct'] = False
        
        return state
        
    except Exception as e:
        display(Markdown(f"**Error:** <span style='color:red'>‡ªÄ‡∫Å‡∫µ‡∫î‡∫Ç‡ªç‡ªâ‡∫ú‡∫¥‡∫î‡∫û‡∫≤‡∫î‡ªÉ‡∫ô‡∫Å‡∫≤‡∫ô‡∫Å‡∫ß‡∫î‡∫™‡∫≠‡∫ö‡∫Ñ‡∫≥‡∫ï‡∫≠‡∫ö: {str(e)}</span>"))
        state['is_correct'] = False
        state['all_correct'] = False
        state['feedback'] = "Unable to check answer due to error"
        return state

# ‡∫™‡ªâ‡∫≤‡∫á Node ‡∫™‡∫∞‡ªÅ‡∫î‡∫á Feedback
def provide_feedback(state: AgentState):
    if not state.get('all_correct', False):
        feedback_md = f"""
**üìù ‡∫ú‡∫ª‡∫ô‡∫Å‡∫≤‡∫ô‡∫Å‡∫ß‡∫î‡∫™‡∫≠‡∫ö:**  
{state.get('feedback', '‡∫Å‡∫∞‡∫•‡∫∏‡∫ô‡∫≤‡∫Å‡∫ß‡∫î‡∫™‡∫≠‡∫ö‡∫Ñ‡∫≥‡∫ï‡∫≠‡∫ö‡∫Ç‡∫≠‡∫á‡∫ó‡ªà‡∫≤‡∫ô')}  
        """
        display(Markdown(feedback_md))
    return state

# Define the conditional routing function
def decide_next_step(state: AgentState):
    """Decide whether to end or provide feedback based on answer completeness and correctness"""
    # ‡∫ñ‡ªâ‡∫≤‡∫ï‡∫≠‡∫ö‡∫Ñ‡∫ª‡∫ö‡∫ó‡∫∏‡∫Å‡∫Ç‡ªç‡ªâ‡ªÅ‡∫•‡ªâ‡∫ß
    if len(state.get('user_answers', [])) >= state.get('total_questions', 5):
        if state.get('all_correct', False):
            return END
        else:
            return "provide_feedback"
    else:
        # ‡∫ç‡∫±‡∫á‡∫ï‡∫≠‡∫ö‡∫ö‡ªç‡ªà‡∫Ñ‡∫ª‡∫ö - ‡∫ï‡ªâ‡∫≠‡∫á‡∫Æ‡ªç‡ªâ input ‡∫à‡∫≤‡∫Å‡∫ú‡∫π‡ªâ‡ªÉ‡∫ä‡ªâ
        return END

# Create the workflow graph
workflow = StateGraph(AgentState)

# Add all nodes
workflow.add_node("web_search", web_search)
workflow.add_node("create_content", create_reading_content) 
workflow.add_node("check_answer", check_answer)
workflow.add_node("provide_feedback", provide_feedback)

# Set up the main flow
workflow.add_edge("web_search", "create_content")
workflow.add_edge("create_content", "check_answer")

# Add conditional routing after check_answer
workflow.add_conditional_edges(
    "check_answer",
    decide_next_step,
    {
        "provide_feedback": "provide_feedback",
        END: END
    }
)

# Add edge from provide_feedback to END to prevent recursion
workflow.add_edge("provide_feedback", END)

# Set entry point
workflow.set_entry_point("web_search")

# Compile the app
app = workflow.compile()

# ‡∫™‡∫∞‡ªÅ‡∫î‡∫á workflow graph (‡∫ñ‡ªâ‡∫≤‡∫ï‡ªâ‡∫≠‡∫á‡∫Å‡∫≤‡∫ô)
try:
    display(Image(app.get_graph().draw_mermaid_png()))
except Exception as e:
    display(Markdown(f"**Note:** Unable to display graph: {str(e)}"))

# ‡∫ü‡∫±‡∫á‡∫ä‡∫±‡∫ô‡∫™‡∫≥‡∫´‡∫º‡∫±‡∫ö‡∫Å‡∫≤‡∫ô‡ªÉ‡∫ä‡ªâ‡∫á‡∫≤‡∫ô
def run_ielts_reading_agent(topic: str):
    # ‡ªÄ‡∫•‡∫µ‡ªà‡∫°‡∫ï‡∫ª‡ªâ‡∫ô‡∫Å‡∫≤‡∫ô‡ªÉ‡∫ä‡ªâ‡∫á‡∫≤‡∫ô
    initial_state = AgentState(
        question=topic,
        search_results=[],
        content="",
        user_answers=[],
        current_question_num=1,
        total_questions=5,
        feedback="",
        is_correct=False,
        all_correct=False
    )
    
    # ‡∫î‡∫∂‡∫á‡∫Ç‡ªç‡ªâ‡∫°‡∫π‡∫ô ‡ªÅ‡∫•‡∫∞ ‡∫™‡ªâ‡∫≤‡∫á‡∫ö‡∫ª‡∫î‡∫Ñ‡∫ß‡∫≤‡∫°
    result = app.invoke(initial_state)
    display(Markdown("\n**--- ‡∫ö‡∫ª‡∫î‡∫Ñ‡∫ß‡∫≤‡∫° Reading IELTS ---**"))
    display(Markdown(result['content']))
    display(Markdown("\n**‡∫Å‡∫∞‡∫•‡∫∏‡∫ô‡∫≤‡∫ï‡∫≠‡∫ö‡∫Ñ‡∫≥‡∫ñ‡∫≤‡∫°‡∫ï‡∫≤‡∫°‡∫•‡∫≥‡∫î‡∫±‡∫ö (1, 2, 3, 4, 5):**"))
    
    # ‡∫Æ‡∫±‡∫ö‡∫Ñ‡∫≥‡∫ï‡∫≠‡∫ö‡∫à‡∫≤‡∫Å‡∫ú‡∫π‡ªâ‡ªÉ‡∫ä‡ªâ‡ªÅ‡∫ö‡∫ö‡∫•‡∫≥‡∫î‡∫±‡∫ö
    question_num = 1
    user_answers = []
    
    # ‡∫Æ‡∫±‡∫ö‡∫Ñ‡∫≥‡∫ï‡∫≠‡∫ö‡∫ó‡∫∏‡∫Å‡∫Ç‡ªç‡ªâ‡∫Å‡ªà‡∫≠‡∫ô
    while question_num <= 5:
        user_answer = input(f"\n‡∫Ñ‡∫≥‡∫ñ‡∫≤‡∫°‡∫Ç‡ªç‡ªâ‡∫ó‡∫µ‡ªà {question_num}: ")
        
        if user_answer.lower() == 'exit':
            return
            
        user_answers.append(user_answer)
        question_num += 1
    
    # ‡∫™‡∫ª‡ªà‡∫á‡∫Ñ‡∫≥‡∫ï‡∫≠‡∫ö‡∫ó‡∫±‡∫á‡ªù‡∫ª‡∫î‡ªÑ‡∫õ‡∫Å‡∫ß‡∫î‡∫™‡∫≠‡∫ö
    result['user_answers'] = user_answers
    result['current_question_num'] = 6  # ‡∫ï‡∫≠‡∫ö‡∫Ñ‡∫ª‡∫ö‡ªÅ‡∫•‡ªâ‡∫ß
    result = app.invoke(result)
    
    if result.get('all_correct', False):
        display(Markdown("\n**üéâ ‡∫ç‡∫¥‡∫ô‡∫î‡∫µ‡∫î‡ªâ‡∫ß‡∫ç! ‡∫ó‡ªà‡∫≤‡∫ô‡∫ï‡∫≠‡∫ö‡∫ñ‡∫∑‡∫Å‡∫ï‡ªâ‡∫≠‡∫á‡∫ó‡∫∏‡∫Å‡∫Ç‡ªç‡ªâ‡ªÅ‡∫•‡ªâ‡∫ß! üéâ**"))
    else:
        display(Markdown("\n**‡∫Å‡∫∞‡∫•‡∫∏‡∫ô‡∫≤‡∫Å‡∫ß‡∫î‡ªÄ‡∫ö‡∫¥‡ªà‡∫á feedback ‡∫Ç‡ªâ‡∫≤‡∫á‡ªÄ‡∫ó‡∫¥‡∫á ‡ªÅ‡∫•‡∫∞ ‡∫•‡∫≠‡∫á‡ªÉ‡ªù‡ªà‡∫≠‡∫µ‡∫Å‡∫Ñ‡∫±‡ªâ‡∫á**"))

# ‡∫ï‡∫ª‡∫ß‡∫¢‡ªà‡∫≤‡∫á‡∫Å‡∫≤‡∫ô‡ªÉ‡∫ä‡ªâ‡∫á‡∫≤‡∫ô
if __name__ == "__main__":
    topic = input("‡∫Å‡∫∞‡∫•‡∫∏‡∫ô‡∫≤‡∫õ‡ªâ‡∫≠‡∫ô‡∫´‡∫ª‡∫ß‡∫Ç‡ªç‡ªâ‡∫ó‡∫µ‡ªà‡∫ó‡ªà‡∫≤‡∫ô‡∫ï‡ªâ‡∫≠‡∫á‡∫Å‡∫≤‡∫ô‡∫ù‡∫∂‡∫Å‡∫≠‡∫≠‡∫Å: ")
    run_ielts_reading_agent(topic)
```
