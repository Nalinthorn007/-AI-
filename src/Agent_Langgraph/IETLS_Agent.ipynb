```python
import os
from typing import TypedDict
from google.oauth2 import service_account
from langchain_community.tools.tavily_search import TavilySearchResults
from langchain_google_vertexai import ChatVertexAI
from IPython.display import Image, display, Markdown
from langgraph.graph import StateGraph, END

# ຕັ້ງຄ່າ API Keys
TAVILY_API_KEY = os.getenv("TAVILY_API_KEY")  # ປ່ຽນເປັນ API Key ຈິງຂອງທ່ານ

# ຕັ້ງຄ່າ Google Vertex AI
JSON_KEY_PATH = r"C:\Users\Kenshin\Desktop\GoogleI-O\key.json"
PROJECT_ID = os.getenv("GOOGLE_CLOUD_PROJECT")
GOOGLE_LOCATION = os.getenv("GOOGLE_LOCATION")

credentials = service_account.Credentials.from_service_account_file(
    JSON_KEY_PATH,
    scopes=["https://www.googleapis.com/auth/cloud-platform"]
)

# ເລືອກໂມເດລ Gemini
model = ChatVertexAI(
    model_name="gemini-2.5-flash-lite",
    project=PROJECT_ID,
    location=GOOGLE_LOCATION,
    credentials=credentials,
    temperature=1.0
)

class AgentState(TypedDict):
    question: str
    search_results: list
    content: str
    user_answers: list
    current_question_num: int
    total_questions: int
    feedback: str
    is_correct: bool
    all_correct: bool

# ສ້າງ Node ຄົ້ນຫາຈາກອິນເຕີເນັດ (5 ແຫຼ່ງ)
def web_search(state: AgentState):
    try:
        # ສ້າງ Tavily search tool
        search = TavilySearchResults(api_key=TAVILY_API_KEY)
        
        # ຄົ້ນຫາຂໍ້ມູນທີ່ກ່ຽວຂ້ອງ
        results = search.invoke({
            "query": f"IELTS reading topic {state['question']} site:cambridge.org OR site:ielts.org",
            "max_results": 5  # ຈຳກັດພຽງ 5 ຜົນການຄົ້ນຫາ
        })
        
        state['search_results'] = results
        return state
    
    except Exception as e:
        display(Markdown(f"**Error:** <span style='color:red'>ເກີດຂໍ້ຜິດພາດໃນການຄົ້ນຫາ: {str(e)}</span>"))
        state['search_results'] = []
        return state

# ສ້າງ Node ສ້າງບົດຄວາມ
def create_reading_content(state: AgentState):
    try:
        if not state['search_results']:
            raise ValueError("ບໍ່ມີຜົນການຄົ້ນຫາ")
            
        # ສ້າງຄຳອະທິບາຍຈາກຜົນການຄົ້ນຫາ
        sources_text = "\n\n".join([
            f"ແຫຼ່ງທີ່ {i+1}:\nURL: {result['url']}\nເນື້ອຫາ: {result['content'][:200]}..." 
            for i, result in enumerate(state['search_results'])
        ])
        
        # ໃຊ້ LLM ສ້າງບົດຄວາມ Reading IELTS
        prompt = f"""
        ຈາກຜົນການຄົ້ນຫາ 5 ແຫຼ່ງຕໍ່ໄປນີ້ ກ່ຽວກັບຫົວຂໍ້ '{state['question']}', 
        ກະລຸນາສ້າງບົດຄວາມ Reading IELTS ທີ່ເໝາະສົມ:

        {sources_text}

        ຂໍ້ກຳນົດ:
        1. ຂຽນເປັນພາສາອັງກິດ ລະດັບ IELTS Academic
        2. ຍາວປະມານ 300-400 ຄຳ
        3. ສ້າງຄຳຖາມພຽງ 2 ປະເພດເທົ່ານັ້ນ:
           - **Sentence Completion** (3 ຂໍ້): ສ້າງປະໂຫຍກທີ່ບໍ່ສົມບູນ ແລະ ໃຫ້ຜູ້ໃຊ້ຕື່ມຄຳຈາກບົດຄວາມ (ບໍ່ເກີນ 3 ຄຳ)
           - **True/False/Not Given** (2 ຂໍ້): ສ້າງປະໂຫຍກເພື່ອໃຫ້ຜູ້ໃຊ້ກວດສອບວ່າ True, False, ຫຼື Not Given
        4. ເນັ້ນການນຳໃຊ້ຂໍ້ມູນຈາກແຫຼ່ງທີ່ນໍາເຊື່ອຖືໄດ້
        
        **ສຳຄັນ: ແຍກຜົນໄນໃຫ້ຊັດເຈນ**
        
        ຈັດຮູບແບບດັ່ງນີ້:
        
        === READING PASSAGE ===
        [ບົດຄວາມ 300-400 ຄຳ]
        
        === QUESTIONS ===
        **Sentence Completion (Choose NO MORE THAN THREE WORDS from the passage):**
        1. The research shows that _______ is the main cause of...
        2. Scientists discovered that _______ can improve...
        
        **True/False/Not Given:**
        3. The study was conducted over a period of five years.
        4. All participants showed positive results.
        5. The research was funded by government agencies.
        
        === ANSWER_KEY_INTERNAL ===
        1. [exact words from passage]
        2. [exact words from passage]  
        3. True/False/Not Given
        4. True/False/Not Given
        5. True/False/Not Given
        """
        
        response = model.invoke(prompt)
        full_content = response.content
        
        # ແຍກເນື້ອຫາອອກຈາກ Answer Key
        if "=== ANSWER_KEY_INTERNAL ===" in full_content:
            content_parts = full_content.split("=== ANSWER_KEY_INTERNAL ===")
            # ເອົາເຉພາະສ່ວນທີ່ຈະສະແດງໃຫ້ຜູ້ໃຊ້
            state['content'] = content_parts[0].replace("=== READING PASSAGE ===", "**Reading Passage:**").replace("=== QUESTIONS ===", "**Questions:**").strip()
            # ເກັບ Answer Key ໄວ້ພາຍໃນ
            state['answer_key'] = content_parts[1].strip() if len(content_parts) > 1 else ""
        else:
            state['content'] = full_content
            state['answer_key'] = ""
        return state
    
    except Exception as e:
        display(Markdown(f"**Error:** <span style='color:red'>ເກີດຂໍ້ຜິດພາດໃນການສ້າງບົດຄວາມ: {str(e)}</span>"))
        state['content'] = ""
        return state

# ສ້າງ Node ກວດສອບຄຳຕອບ
def check_answer(state: AgentState):
    try:
        # ກວດສອບຄຳຕອບທັງໝົດ
        if 'user_answers' in state and len(state['user_answers']) >= 5:
            prompt = f"""
            ບົດຄວາມ Reading:
            {state['content']}
            
            ຄຳຕອບຂອງນັກຮຽນ (ຕາມລຳດັບ 1-5):
            {', '.join([f"{i+1}. {ans}" for i, ans in enumerate(state['user_answers'])])}
            
            ກະລຸນາກວດສອບຄຳຕອບທັງໝົດຕາມເງື່ອນໄຂ IELTS Reading:
            
            **ສຳຫຼັບ Sentence Completion:**
            - ກວດສອບວ່າຄຳຕອບມີບໍ່ເກີນ 3 ຄຳ
            - ຄຳຕອບຕ້ອງມາຈາກບົດຄວາມໂດຍກົງ
            - ຕ້ອງຖືກຕ້ອງທາງດ້ານໄວຍາກອນ
            
            **ສຳຫຼັບ True/False/Not Given:**
            - True: ຖ້າຂໍ້ຄວາມກົງກັບບົດຄວາມ
            - False: ຖ້າຂໍ້ຄວາມກົງກັນຂ້າມກັບບົດຄວາມ  
            - Not Given: ຖ້າບໍ່ມີຂໍ້ມູນໃນບົດຄວາມ
            
            ກະລຸນາໃຫ້ຜົນການກວດສອບແຕ່ລະຂໍ້:
            1. ຖືກ/ຜິດ ແລະ ເຫດຜົນ
            2. ຖືກ/ຜິດ ແລະ ເຫດຜົນ
            3. ຖືກ/ຜິດ ແລະ ເຫດຜົນ
            4. ຖືກ/ຜິດ ແລະ ເຫດຜົນ
            5. ຖືກ/ຜິດ ແລະ ເຫດຜົນ
            
            ຖ້າມີຂໍ້ໃດຜິດ, ໃຫ້ Feedback ແລະ Highlight ສ່ວນທີ່ກ່ຽວຂ້ອງ.
            ຖ້າທຸກຂໍ້ຖືກຕ້ອງ, ໃຫ້ຕອບວ່າ "ALL CORRECT" ຢ່າງຊັດເຈນ.
            """
            
            response = model.invoke(prompt)
            feedback = response.content
            
            # ກວດສອບວ່າທຸກຂໍ້ຖືກຕ້ອງຫຼືບໍ່
            if "ALL CORRECT" in feedback.upper():
                state['all_correct'] = True
                state['is_correct'] = True
            else:
                state['all_correct'] = False
                state['is_correct'] = False
                state['feedback'] = feedback
        else:
            # ຍັງຕອບບໍ່ຄົບ
            state['is_correct'] = False
            state['all_correct'] = False
        
        return state
        
    except Exception as e:
        display(Markdown(f"**Error:** <span style='color:red'>ເກີດຂໍ້ຜິດພາດໃນການກວດສອບຄຳຕອບ: {str(e)}</span>"))
        state['is_correct'] = False
        state['all_correct'] = False
        state['feedback'] = "Unable to check answer due to error"
        return state

# ສ້າງ Node ສະແດງ Feedback
def provide_feedback(state: AgentState):
    if not state.get('all_correct', False):
        feedback_md = f"""
**📝 ຜົນການກວດສອບ:**  
{state.get('feedback', 'ກະລຸນາກວດສອບຄຳຕອບຂອງທ່ານ')}  
        """
        display(Markdown(feedback_md))
    return state

# Define the conditional routing function
def decide_next_step(state: AgentState):
    """Decide whether to end or provide feedback based on answer completeness and correctness"""
    # ຖ້າຕອບຄົບທຸກຂໍ້ແລ້ວ
    if len(state.get('user_answers', [])) >= state.get('total_questions', 5):
        if state.get('all_correct', False):
            return END
        else:
            return "provide_feedback"
    else:
        # ຍັງຕອບບໍ່ຄົບ - ຕ້ອງຮໍ້ input ຈາກຜູ້ໃຊ້
        return END

# Create the workflow graph
workflow = StateGraph(AgentState)

# Add all nodes
workflow.add_node("web_search", web_search)
workflow.add_node("create_content", create_reading_content) 
workflow.add_node("check_answer", check_answer)
workflow.add_node("provide_feedback", provide_feedback)

# Set up the main flow
workflow.add_edge("web_search", "create_content")
workflow.add_edge("create_content", "check_answer")

# Add conditional routing after check_answer
workflow.add_conditional_edges(
    "check_answer",
    decide_next_step,
    {
        "provide_feedback": "provide_feedback",
        END: END
    }
)

# Add edge from provide_feedback to END to prevent recursion
workflow.add_edge("provide_feedback", END)

# Set entry point
workflow.set_entry_point("web_search")

# Compile the app
app = workflow.compile()

# ສະແດງ workflow graph (ຖ້າຕ້ອງການ)
try:
    display(Image(app.get_graph().draw_mermaid_png()))
except Exception as e:
    display(Markdown(f"**Note:** Unable to display graph: {str(e)}"))

# ຟັງຊັນສຳຫຼັບການໃຊ້ງານ
def run_ielts_reading_agent(topic: str):
    # ເລີ່ມຕົ້ນການໃຊ້ງານ
    initial_state = AgentState(
        question=topic,
        search_results=[],
        content="",
        user_answers=[],
        current_question_num=1,
        total_questions=5,
        feedback="",
        is_correct=False,
        all_correct=False
    )
    
    # ດຶງຂໍ້ມູນ ແລະ ສ້າງບົດຄວາມ
    result = app.invoke(initial_state)
    display(Markdown("\n**--- ບົດຄວາມ Reading IELTS ---**"))
    display(Markdown(result['content']))
    display(Markdown("\n**ກະລຸນາຕອບຄຳຖາມຕາມລຳດັບ (1, 2, 3, 4, 5):**"))
    
    # ຮັບຄຳຕອບຈາກຜູ້ໃຊ້ແບບລຳດັບ
    question_num = 1
    user_answers = []
    
    # ຮັບຄຳຕອບທຸກຂໍ້ກ່ອນ
    while question_num <= 5:
        user_answer = input(f"\nຄຳຖາມຂໍ້ທີ່ {question_num}: ")
        
        if user_answer.lower() == 'exit':
            return
            
        user_answers.append(user_answer)
        question_num += 1
    
    # ສົ່ງຄຳຕອບທັງໝົດໄປກວດສອບ
    result['user_answers'] = user_answers
    result['current_question_num'] = 6  # ຕອບຄົບແລ້ວ
    result = app.invoke(result)
    
    if result.get('all_correct', False):
        display(Markdown("\n**🎉 ຍິນດີດ້ວຍ! ທ່ານຕອບຖືກຕ້ອງທຸກຂໍ້ແລ້ວ! 🎉**"))
    else:
        display(Markdown("\n**ກະລຸນາກວດເບິ່ງ feedback ຂ້າງເທິງ ແລະ ລອງໃໝ່ອີກຄັ້ງ**"))

# ຕົວຢ່າງການໃຊ້ງານ
if __name__ == "__main__":
    topic = input("ກະລຸນາປ້ອນຫົວຂໍ້ທີ່ທ່ານຕ້ອງການຝຶກອອກ: ")
    run_ielts_reading_agent(topic)
```
