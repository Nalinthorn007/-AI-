```python
import os
import json
import base64
from typing import TypedDict, List, Dict, Any
from IPython.display import display, Markdown, Image
import requests
from PIL import Image as PILImage

# Langgraph imports
from langgraph.graph import StateGraph
from langchain_core.messages import BaseMessage, HumanMessage, AIMessage, ToolMessage
from langgraph.checkpoint.memory import MemorySaver

# LLM imports
from langchain_groq import ChatGroq
import openai

# Tools
from langchain_core.tools import tool

# ‡∫Å‡∫≤‡∫ô‡∫ë‡∫±‡ªâ‡∫á‡∫Ñ‡ªà‡∫≤ API
GROQ_API_KEY = os.getenv("GROQ_API_KEY")
SERPAPI_API_KEY = os.getenv("SERPAPI_API_KEY")
OPENROUTER_API_KEY = os.getenv("OPENROUTER_API_KEY")

# ‡∫ï‡∫±‡ªâ‡∫á‡∫Ñ‡ªà‡∫≤ APIs
os.environ["GROQ_API_KEY"] = GROQ_API_KEY

class AgentState(TypedDict):
    messages: List[BaseMessage]
    next_action: str

# ‡ªÄ‡∫•‡∫µ‡ªà‡∫°‡∫ï‡∫ª‡ªâ‡∫ô LLMs
llm = ChatGroq(
    model="meta-llama/llama-4-maverick-17b-128e-instruct",
    temperature=0,
    max_tokens=1024
)

# Vision LLM ‡∫™‡∫≥‡∫•‡∫±‡∫ö‡∫Å‡∫≤‡∫ô‡∫õ‡∫∞‡∫°‡∫ß‡∫ô‡∫ú‡∫ª‡∫ô‡∫Æ‡∫π‡∫ö‡∫û‡∫≤‡∫ö
vision_client = openai.OpenAI(
    base_url="https://openrouter.ai/api/v1",
    api_key=OPENROUTER_API_KEY,
)

@tool
def search_flights(departure: str, arrival: str, date: str, return_date: str = None, trip_type: str = "one-way", airline_filter: str = None) -> Dict[str, Any]:
    """
    ‡∫Ñ‡∫ª‡ªâ‡∫ô‡∫´‡∫≤‡∫ó‡ªà‡∫Ω‡∫ß‡∫ö‡∫¥‡∫ô‡ªÇ‡∫î‡∫ç‡ªÉ‡∫ä‡ªâ SerpAPI Google Flights
    
    Args:
        departure: ‡∫•‡∫∞‡∫´‡∫±‡∫î‡∫™‡∫∞‡ªú‡∫≤‡∫°‡∫ö‡∫¥‡∫ô ‡∫´‡∫º‡∫∑ ‡ªÄ‡∫°‡∫∑‡∫≠‡∫á‡∫ï‡∫ª‡ªâ‡∫ô‡∫ó‡∫≤‡∫á
        arrival: ‡∫•‡∫∞‡∫´‡∫±‡∫î‡∫™‡∫∞‡ªú‡∫≤‡∫°‡∫ö‡∫¥‡∫ô ‡∫´‡∫º‡∫∑ ‡ªÄ‡∫°‡∫∑‡∫≠‡∫á‡∫õ‡∫≤‡∫ç‡∫ó‡∫≤‡∫á  
        date: ‡∫ß‡∫±‡∫ô‡∫ó‡∫µ‡∫ó‡ªà‡∫Ω‡∫ß‡∫ö‡∫¥‡∫ô‡ªÑ‡∫õ‡ªÉ‡∫ô‡∫Æ‡∫π‡∫ö‡ªÅ‡∫ö‡∫ö YYYY-MM-DD
        return_date: ‡∫ß‡∫±‡∫ô‡∫ó‡∫µ‡∫ó‡ªà‡∫Ω‡∫ß‡∫ö‡∫¥‡∫ô‡∫Å‡∫±‡∫ö‡ªÉ‡∫ô‡∫Æ‡∫π‡∫ö‡ªÅ‡∫ö‡∫ö YYYY-MM-DD (‡∫™‡∫≥‡∫•‡∫±‡∫ö round trip)
        trip_type: ‡∫õ‡∫∞‡ªÄ‡∫û‡∫î‡∫Å‡∫≤‡∫ô‡ªÄ‡∫î‡∫µ‡∫ô‡∫ó‡∫≤‡∫á "one-way" ‡∫´‡∫º‡∫∑ "round-trip"
        airline_filter: ‡∫ü‡∫¥‡∫ß‡ªÄ‡∫ï‡∫µ‡ªâ‡∫™‡∫≤‡∫ç‡∫Å‡∫≤‡∫ô‡∫ö‡∫¥‡∫ô (‡ªÄ‡∫ä‡∫±‡ªà‡∫ô: "AirAsia", "Thai", "Lao Airlines")
    """
    try:
        url = "https://serpapi.com/search.json"
        
        # ‡∫Å‡∫≥‡∫ô‡∫ª‡∫î type: 2 = one-way, 1 = round-trip
        flight_type = 2 if trip_type.lower() == "one-way" else 1
        
        params = {
            "engine": "google_flights",
            "departure_id": departure,
            "arrival_id": arrival,
            "outbound_date": date,
            "type": flight_type,
            "currency": "USD",
            "hl": "lo",
            "api_key": SERPAPI_API_KEY
        }
        
        # ‡ªÄ‡∫û‡∫µ‡ªà‡∫° return_date ‡∫™‡∫≥‡∫•‡∫±‡∫ö round trip
        if flight_type == 1 and return_date:
            params["return_date"] = return_date
        elif flight_type == 1 and not return_date:
            # ‡∫ñ‡ªâ‡∫≤‡∫ö‡ªç‡ªà‡ªÑ‡∫î‡ªâ‡∫•‡∫∞‡∫ö‡∫∏‡∫ß‡∫±‡∫ô‡∫Å‡∫±‡∫ö, ‡ªÉ‡∫´‡ªâ‡∫Ñ‡∫ª‡ªâ‡∫ô‡∫´‡∫≤‡ªÅ‡∫ö‡∫ö one-way ‡ªÅ‡∫ó‡∫ô
            params["type"] = 2
            print("‚ö†Ô∏è ‡∫ö‡ªç‡ªà‡ªÑ‡∫î‡ªâ‡∫•‡∫∞‡∫ö‡∫∏‡∫ß‡∫±‡∫ô‡∫ó‡∫µ‡∫Å‡∫±‡∫ö, ‡∫Ñ‡∫ª‡ªâ‡∫ô‡∫´‡∫≤‡ªÅ‡∫ö‡∫ö one-way")
        
        filter_msg = f" (‡∫™‡∫≤‡∫ç‡∫Å‡∫≤‡∫ô‡∫ö‡∫¥‡∫ô: {airline_filter})" if airline_filter else ""
        print(f"üîç ‡∫Ñ‡∫ª‡ªâ‡∫ô‡∫´‡∫≤‡∫ó‡ªà‡∫Ω‡∫ß‡∫ö‡∫¥‡∫ô: {departure} ‚Üí {arrival}, {date}" + 
              (f" ‚Üí {return_date}" if return_date else " (‡∫ó‡∫≤‡∫á‡∫î‡∫Ω‡∫ß)") + filter_msg)
        
        response = requests.get(url, params=params)
        data = response.json()
        
        # ‡∫Å‡∫ß‡∫î‡∫™‡∫≠‡∫ö‡∫Ç‡ªç‡ªâ‡∫ú‡∫¥‡∫î‡∫û‡∫≤‡∫î‡∫à‡∫≤‡∫Å API
        if "error" in data:
            return {"error": f"SerpAPI Error: {data['error']}", "status": "‡∫ú‡∫¥‡∫î‡∫û‡∫≤‡∫î"}
        
        # ‡ªÉ‡∫ä‡ªâ other_flights ‡∫´‡∫º‡∫∑ best_flights ‡∫Ç‡∫∂‡ªâ‡∫ô‡∫Å‡∫±‡∫ö‡∫ß‡ªà‡∫≤‡∫°‡∫µ‡∫Ç‡ªç‡ªâ‡∫°‡∫π‡∫ô‡∫≠‡∫±‡∫ô‡ªÉ‡∫î
        flights_data = data.get("best_flights", data.get("other_flights", []))
        
        if flights_data:
            flights = []
            filtered_count = 0
            
            for flight in flights_data:
                # ‡∫à‡∫±‡∫î‡∫Å‡∫≤‡∫ô‡∫Ç‡ªç‡ªâ‡∫°‡∫π‡∫ô‡∫ó‡ªà‡∫Ω‡∫ß‡∫ö‡∫¥‡∫ô‡∫ü‡∫≤‡∫°‡ªÇ‡∫Ñ‡∫á‡∫™‡ªâ‡∫≤‡∫á‡∫ó‡∫µ‡ªà‡ªÅ‡∫ó‡ªâ‡∫à‡∫¥‡∫á
                flight_legs = flight.get("flights", [])
                if flight_legs:
                    first_leg = flight_legs[0]
                    flight_airline = first_leg.get("airline", "")
                    
                    # ‡∫ü‡∫¥‡∫ß‡ªÄ‡∫ï‡∫µ‡ªâ‡∫™‡∫≤‡∫ç‡∫Å‡∫≤‡∫ô‡∫ö‡∫¥‡∫ô‡∫ñ‡ªâ‡∫≤‡∫°‡∫µ‡∫Å‡∫≤‡∫ô‡∫•‡∫∞‡∫ö‡∫∏
                    if airline_filter:
                        airline_keywords = {
                            "airasia": ["AirAsia", "Air Asia", "AK", "FD", "QZ", "Z2"],
                            "thai": ["THAI", "Thai Airways", "TG"],
                            "lao": ["Lao Airlines", "QV"],
                            "bangkok": ["Bangkok Airways", "PG"],
                            "jetstar": ["Jetstar", "JQ", "BL"],
                            "vietjet": ["VietJet", "VJ"],
                            "cebu": ["Cebu Pacific", "5J"]
                        }
                        
                        airline_lower = airline_filter.lower()
                        should_include = False
                        
                        for key, keywords in airline_keywords.items():
                            if key in airline_lower:
                                for keyword in keywords:
                                    if keyword.lower() in flight_airline.lower():
                                        should_include = True
                                        break
                                break
                        
                        # ‡∫ñ‡ªâ‡∫≤‡∫ö‡ªç‡ªà‡∫û‡∫ª‡∫ö‡ªÉ‡∫ô keywords ‡ªÉ‡∫´‡ªâ‡ªÄ‡∫ä‡∫±‡∫Å‡ªÇ‡∫î‡∫ç‡∫Å‡∫ª‡∫á
                        if not should_include and airline_filter.lower() in flight_airline.lower():
                            should_include = True
                            
                        if not should_include:
                            filtered_count += 1
                            continue
                    
                    flight_info = {
                        "airline": flight_airline,
                        "flight_number": first_leg.get("flight_number", ""),
                        "airplane": first_leg.get("airplane", ""),
                        "departure_airport": {
                            "code": first_leg.get("departure_airport", {}).get("id", ""),
                            "name": first_leg.get("departure_airport", {}).get("name", ""),
                            "time": first_leg.get("departure_airport", {}).get("time", "")
                        },
                        "arrival_airport": {
                            "code": first_leg.get("arrival_airport", {}).get("id", ""),
                            "name": first_leg.get("arrival_airport", {}).get("name", ""),
                            "time": first_leg.get("arrival_airport", {}).get("time", "")
                        },
                        "duration": f"{flight.get('total_duration', 0)} ‡∫ô‡∫≤‡∫ó‡∫µ",
                        "price": f"${flight.get('price', 0)}",
                        "type": flight.get("type", "One way"),
                        "travel_class": first_leg.get("travel_class", "Economy"),
                        "legroom": first_leg.get("legroom", ""),
                        "stops": len(flight_legs) - 1,
                        "carbon_emissions": flight.get("carbon_emissions", {}).get("this_flight", 0),
                        "booking_token": flight.get("booking_token", ""),
                        "extensions": first_leg.get("extensions", [])
                    }
                    
                    # ‡ªÄ‡∫û‡∫µ‡ªà‡∫°‡∫Ç‡ªç‡ªâ‡∫°‡∫π‡∫ô‡∫ó‡ªà‡∫Ω‡∫ß‡∫ö‡∫¥‡∫ô‡∫Å‡∫±‡∫ö‡∫™‡∫≥‡∫•‡∫±‡∫ö round trip
                    if len(flight_legs) > 1:
                        return_leg = flight_legs[-1]  # ‡∫ó‡ªà‡∫Ω‡∫ß‡∫ö‡∫¥‡∫ô‡∫Å‡∫±‡∫ö
                        flight_info["return_flight"] = {
                            "departure_time": return_leg.get("departure_airport", {}).get("time", ""),
                            "arrival_time": return_leg.get("arrival_airport", {}).get("time", ""),
                            "flight_number": return_leg.get("flight_number", ""),
                            "airline": return_leg.get("airline", "")
                        }
                    
                    flights.append(flight_info)
                    
                    # ‡∫à‡∫≥‡∫Å‡∫±‡∫î‡∫ú‡∫ª‡∫ô‡∫•‡∫±‡∫ö‡ªÉ‡∫´‡ªâ‡∫™‡∫π‡∫á‡∫™‡∫∏‡∫î 5 ‡∫ó‡ªà‡∫Ω‡∫ß‡∫ö‡∫¥‡∫ô
                    if len(flights) >= 5:
                        break
            
            # ‡ªÄ‡∫û‡∫µ‡ªà‡∫°‡∫Ç‡ªç‡ªâ‡∫°‡∫π‡∫ô‡∫™‡∫∞‡ªú‡∫≤‡∫°‡∫ö‡∫¥‡∫ô
            airports_info = data.get("airports", [])
            departure_info = {}
            arrival_info = {}
            
            if airports_info:
                if "departure" in airports_info[0]:
                    dep_data = airports_info[0]["departure"][0]
                    departure_info = {
                        "airport_name": dep_data.get("airport", {}).get("name", ""),
                        "city": dep_data.get("city", ""),
                        "country": dep_data.get("country", ""),
                        "code": dep_data.get("airport", {}).get("id", "")
                    }
                
                if "arrival" in airports_info[0]:
                    arr_data = airports_info[0]["arrival"][0]
                    arrival_info = {
                        "airport_name": arr_data.get("airport", {}).get("name", ""),
                        "city": arr_data.get("city", ""),
                        "country": arr_data.get("country", ""),
                        "code": arr_data.get("airport", {}).get("id", "")
                    }
            
            result = {
                "flights": flights,
                "route_info": {
                    "departure": departure_info,
                    "arrival": arrival_info
                },
                "search_params": {
                    "departure": departure,
                    "arrival": arrival,
                    "outbound_date": date,
                    "return_date": return_date,
                    "trip_type": trip_type,
                    "airline_filter": airline_filter
                },
                "total_found": len(flights),
                "status": "‡∫™‡∫≥‡ªÄ‡∫•‡∫±‡∫î"
            }
            
            # ‡ªÄ‡∫û‡∫µ‡ªà‡∫°‡∫Ç‡ªç‡ªâ‡∫Ñ‡∫ß‡∫≤‡∫°‡∫ñ‡ªâ‡∫≤‡∫°‡∫µ‡∫Å‡∫≤‡∫ô‡∫ü‡∫¥‡∫ß‡ªÄ‡∫ï‡∫µ‡ªâ
            if airline_filter and filtered_count > 0:
                result["filter_info"] = {
                    "filtered_out": filtered_count,
                    "message": f"‡∫ü‡∫¥‡∫ß‡ªÄ‡∫ï‡∫µ‡ªâ‡∫≠‡∫≠‡∫Å {filtered_count} ‡∫ó‡ªà‡∫Ω‡∫ß‡∫ö‡∫¥‡∫ô‡∫ó‡∫µ‡ªà‡∫ö‡ªç‡ªà‡ªÅ‡∫°‡ªà‡∫ô {airline_filter}"
                }
            
            return result
        else:
            return {"error": "‡∫ö‡ªç‡ªà‡∫û‡∫ª‡∫ö‡∫ó‡ªà‡∫Ω‡∫ß‡∫ö‡∫¥‡∫ô‡∫ó‡∫µ‡ªà‡∫ï‡∫ª‡∫á‡∫Å‡∫±‡∫ö‡ªÄ‡∫á‡∫∑‡ªà‡∫≠‡∫ô‡ªÑ‡∫Ç", "status": "‡∫ú‡∫¥‡∫î‡∫û‡∫≤‡∫î"}
            
    except requests.exceptions.RequestException as e:
        return {"error": f"‡∫Ç‡ªç‡ªâ‡∫ú‡∫¥‡∫î‡∫û‡∫≤‡∫î‡∫Å‡∫≤‡∫ô‡ªÄ‡∫ä‡∫∑‡ªà‡∫≠‡∫°‡∫ï‡ªç‡ªà: {str(e)}", "status": "‡∫ú‡∫¥‡∫î‡∫û‡∫≤‡∫î"}
    except Exception as e:
        return {"error": f"‡∫Ç‡ªç‡ªâ‡∫ú‡∫¥‡∫î‡∫û‡∫≤‡∫î‡∫ó‡∫ª‡ªà‡∫ß‡ªÑ‡∫õ: {str(e)}", "status": "‡∫ú‡∫¥‡∫î‡∫û‡∫≤‡∫î"}

@tool  
def analyze_booking_images() -> Dict[str, Any]:
    """
    ‡∫ß‡∫¥‡ªÄ‡∫Ñ‡∫≤‡∫∞‡∫Æ‡∫π‡∫ö‡∫û‡∫≤‡∫ö‡∫Å‡∫≤‡∫ô‡∫à‡∫≠‡∫á‡∫ó‡∫µ‡ªà‡∫¢‡∫∑‡∫ô‡∫¢‡∫±‡∫ô‡∫à‡∫≤‡∫Å‡ªÇ‡∫ü‡∫•‡ªÄ‡∫î‡∫µ‡∫ó‡∫µ‡ªà‡∫Å‡ªç‡∫≤‡ªú‡∫ª‡∫î
    """
    try:
        image_folder = r"D:\Code AI Before Sleep\src\image"
        image_files = ["1.jpg", "2.jpg", "3.jpg"]
        
        all_booking_info = []
        
        for filename in image_files:
            image_path = os.path.join(image_folder, filename)
            
            if os.path.exists(image_path):
                # ‡∫≠‡ªà‡∫≤‡∫ô ‡ªÅ‡∫•‡∫∞ ‡ªÄ‡∫Ç‡∫ª‡ªâ‡∫≤‡∫•‡∫∞‡∫´‡∫±‡∫î‡∫Æ‡∫π‡∫ö‡∫û‡∫≤‡∫ö
                with open(image_path, "rb") as image_file:
                    base64_image = base64.b64encode(image_file.read()).decode('utf-8')
                
                # ‡∫ß‡∫¥‡ªÄ‡∫Ñ‡∫≤‡∫∞‡∫Æ‡∫π‡∫ö‡∫û‡∫≤‡∫ö‡∫î‡ªâ‡∫ß‡∫ç Gemini 2.0 Flash
                response = vision_client.chat.completions.create(
                    model="google/gemini-2.0-flash-exp:free",
                    messages=[
                        {
                            "role": "user",
                            "content": [
                                {
                                    "type": "text",
                                    "text": "‡∫Å‡∫∞‡∫•‡∫∏‡∫ô‡∫≤‡ªÅ‡∫ç‡∫Å‡ªÄ‡∫≠‡∫ª‡∫≤‡∫Ç‡ªç‡ªâ‡∫°‡∫π‡∫ô‡∫Å‡∫≤‡∫ô‡∫à‡∫≠‡∫á‡∫ó‡ªà‡∫Ω‡∫ß‡∫ö‡∫¥‡∫ô‡∫ó‡∫±‡∫á‡ªù‡∫ª‡∫î‡∫à‡∫≤‡∫Å‡∫Æ‡∫π‡∫ö‡∫ô‡∫µ‡ªâ‡∫•‡∫ß‡∫°‡∫ó‡∫±‡∫á: ‡ªù‡∫≤‡∫ç‡ªÄ‡∫•‡∫Å‡∫≠‡ªâ‡∫≤‡∫á‡∫≠‡∫µ‡∫á‡∫Å‡∫≤‡∫ô‡∫à‡∫≠‡∫á, ‡∫ä‡∫∑‡ªà‡∫ú‡∫π‡ªâ‡ªÇ‡∫î‡∫ç‡∫™‡∫≤‡∫ô, ‡ªù‡∫≤‡∫ç‡ªÄ‡∫•‡∫Å‡∫ñ‡ªâ‡∫Ω‡∫ß‡∫ö‡∫¥‡∫ô, ‡∫™‡∫≤‡∫ç‡∫Å‡∫≤‡∫ô‡∫ö‡∫¥‡∫ô, ‡∫™‡∫∞‡ªú‡∫≤‡∫°‡∫ö‡∫¥‡∫ô‡∫≠‡∫≠‡∫Å/‡∫Æ‡∫≠‡∫î, ‡∫ß‡∫±‡∫ô‡∫ó‡∫µ, ‡ªÄ‡∫ß‡∫•‡∫≤, ‡ªù‡∫≤‡∫ç‡ªÄ‡∫•‡∫Å‡∫ó‡∫µ‡ªà‡∫ô‡∫±‡ªà‡∫á ‡ªÅ‡∫•‡∫∞‡∫•‡∫≤‡∫ç‡∫•‡∫∞‡∫≠‡∫Ω‡∫î‡∫≠‡∫∑‡ªà‡∫ô‡ªÜ‡∫ó‡∫µ‡ªà‡∫Å‡ªà‡∫Ω‡∫ß‡∫Ç‡ªâ‡∫≠‡∫á. ‡∫à‡∫±‡∫î‡∫Æ‡∫π‡∫ö‡ªÅ‡∫ö‡∫ö‡∫Å‡∫≤‡∫ô‡∫ï‡∫≠‡∫ö‡∫Å‡∫±‡∫ö‡ªÄ‡∫õ‡∫±‡∫ô‡∫Ç‡ªç‡ªâ‡∫°‡∫π‡∫ô‡∫ó‡∫µ‡ªà‡∫°‡∫µ‡ªÇ‡∫Ñ‡∫á‡∫™‡ªâ‡∫≤‡∫á."
                                },
                                {
                                    "type": "image_url",
                                    "image_url": {
                                        "url": f"data:image/jpeg;base64,{base64_image}"
                                    }
                                }
                            ]
                        }
                    ],
                    max_tokens=1000
                )
                
                booking_data = {
                    "image_file": filename,
                    "extracted_info": response.choices[0].message.content
                }
                all_booking_info.append(booking_data)
            else:
                all_booking_info.append({
                    "image_file": filename,
                    "error": "‡∫ö‡ªç‡ªà‡∫û‡∫ª‡∫ö‡ªÑ‡∫ü‡∫•‡∫å‡∫Æ‡∫π‡∫ö‡∫û‡∫≤‡∫ö"
                })
        
        return {
            "booking_confirmations": all_booking_info,
            "total_processed": len(all_booking_info),
            "status": "‡∫™‡∫≥‡ªÄ‡∫•‡∫±‡∫î"
        }
        
    except Exception as e:
        return {"error": str(e), "status": "‡∫ú‡∫¥‡∫î‡∫û‡∫≤‡∫î"}

# ‡∫•‡∫≤‡∫ç‡∫Å‡∫≤‡∫ô tools ‡∫ó‡∫±‡∫á‡ªù‡∫ª‡∫î - ‡ªÅ‡∫ö‡∫ö‡∫á‡ªà‡∫≤‡∫ç‡ªÜ
tools = [search_flights, analyze_booking_images]

# Assign tools ‡ªÄ‡∫Ç‡∫ª‡ªâ‡∫≤‡∫Å‡∫±‡∫ö LLM 
llm_with_tools = llm.bind_tools(tools)

# Manual tool execution function (‡∫Ñ‡∫∑‡∫Å‡∫±‡∫ô‡∫Å‡∫±‡∫ö‡∫ï‡∫ª‡∫ß‡∫¢‡ªà‡∫≤‡∫á)
def execute_tools(tool_calls, available_tools):
    """set up tool execution"""
    tools_dict = {tool.name: tool for tool in available_tools}
    results = []
    
    for tool_call in tool_calls:
        tool_name = tool_call["name"]
        tool_args = tool_call["args"]
        tool_id = tool_call["id"]
        
        if tool_name in tools_dict:
            try:
                result = tools_dict[tool_name].invoke(tool_args)
                results.append(
                    ToolMessage(
                        content=str(result),
                        tool_call_id=tool_id
                    )
                )
            except Exception as e:
                results.append(
                    ToolMessage(
                        content=f"Error executing {tool_name}: {str(e)}",
                        tool_call_id=tool_id
                    )
                )
        else:
            results.append(
                ToolMessage(
                    content=f"Tool {tool_name} not found",
                    tool_call_id=tool_id
                )
            )
    
    return results

# Assign tools ‡ªÄ‡∫Ç‡∫ª‡ªâ‡∫≤‡∫Å‡∫±‡∫ö LLM 
llm_with_tools = llm.bind_tools(tools)

def should_continue(state: AgentState) -> str:
    """‡∫Å‡∫≥‡∫ô‡∫ª‡∫î‡∫ß‡ªà‡∫≤‡∫Ñ‡∫ß‡∫ô‡∫™‡∫∑‡∫ö‡∫ï‡ªç‡ªà‡∫Å‡∫≤‡∫ô‡∫ó‡∫≥‡∫á‡∫≤‡∫ô‡∫´‡∫º‡∫∑‡∫¢‡∫∏‡∫î"""
    messages = state["messages"]
    last_message = messages[-1]
    
    # ‡∫ñ‡ªâ‡∫≤‡∫Ç‡ªç‡ªâ‡∫Ñ‡∫ß‡∫≤‡∫°‡∫™‡∫∏‡∫î‡∫ó‡ªâ‡∫≤‡∫ç‡∫ö‡ªç‡ªà‡ªÑ‡∫î‡ªâ‡ªÄ‡∫≠‡∫µ‡ªâ‡∫ô‡ªÉ‡∫ä‡ªâ tool ‡ªÅ‡∫°‡ªà‡∫ô‡∫™‡∫¥‡ªâ‡∫ô‡∫™‡∫∏‡∫î
    if not hasattr(last_message, 'tool_calls') or not last_message.tool_calls:
        return "end"
    else:
        return "continue"

def call_model(state: AgentState):
    """‡ªÄ‡∫≠‡∫µ‡ªâ‡∫ô LLM ‡ªÄ‡∫û‡∫∑‡ªà‡∫≠‡∫™‡ªâ‡∫≤‡∫á‡∫Ñ‡∫≥‡∫ï‡∫≠‡∫ö"""
    messages = state["messages"]
    
    # System prompt ‡∫™‡∫≥‡∫•‡∫±‡∫ö ReAct pattern
    system_prompt = """‡ªÄ‡∫à‡∫ª‡ªâ‡∫≤‡ªÅ‡∫°‡ªà‡∫ô AI Assistant ‡∫™‡∫≤‡∫ç‡∫Å‡∫≤‡∫ô‡∫ö‡∫¥‡∫ô‡∫ó‡∫µ‡ªà‡∫™‡∫≤‡∫°‡∫≤‡∫î‡ªÉ‡∫ä‡ªâ‡ªÄ‡∫Ñ‡∫∑‡ªà‡∫≠‡∫á‡∫°‡∫∑‡∫ï‡ªà‡∫≤‡∫á‡ªÜ‡ªÑ‡∫î‡ªâ.

‡ªÄ‡∫à‡∫ª‡ªâ‡∫≤‡∫°‡∫µ‡ªÄ‡∫Ñ‡∫∑‡ªà‡∫≠‡∫á‡∫°‡∫∑ 2 ‡∫¢‡ªà‡∫≤‡∫á:
1. search_flights - ‡∫™‡∫≥‡∫•‡∫±‡∫ö‡∫Ñ‡∫ª‡ªâ‡∫ô‡∫´‡∫≤‡∫ó‡ªà‡∫Ω‡∫ß‡∫ö‡∫¥‡∫ô (‡∫ï‡ªâ‡∫≠‡∫á‡∫Å‡∫≤‡∫ô departure, arrival, date)
2. analyze_booking_images - ‡∫™‡∫≥‡∫•‡∫±‡∫ö‡∫ß‡∫¥‡ªÄ‡∫Ñ‡∫≤‡∫∞‡∫Æ‡∫π‡∫ö‡∫û‡∫≤‡∫ö‡∫Å‡∫≤‡∫ô‡∫à‡∫≠‡∫á‡∫ó‡∫µ‡ªà‡ªÄ‡∫Å‡∫±‡∫ö‡ªÑ‡∫ß‡ªâ

‡ªÄ‡∫°‡∫∑‡ªà‡∫≠‡∫ú‡∫π‡ªâ‡ªÉ‡∫ä‡ªâ‡∫ñ‡∫≤‡∫°, ‡ªÉ‡∫´‡ªâ‡ªÄ‡∫à‡∫ª‡ªâ‡∫≤‡∫õ‡∫∞‡∫ï‡∫¥‡∫ö‡∫±‡∫î‡∫ï‡∫≤‡∫° ReAct pattern:
1. Reason - ‡∫ß‡∫¥‡ªÄ‡∫Ñ‡∫≤‡∫∞‡∫Ñ‡∫≥‡∫ñ‡∫≤‡∫°‡ªÅ‡∫•‡∫∞‡∫Ñ‡∫¥‡∫î‡∫ß‡ªà‡∫≤‡∫ï‡ªâ‡∫≠‡∫á‡ªÄ‡∫Æ‡∫±‡∫î‡∫´‡∫ç‡∫±‡∫á
2. Act - ‡ªÉ‡∫ä‡ªâ‡ªÄ‡∫Ñ‡∫∑‡ªà‡∫≠‡∫á‡∫°‡∫∑‡∫ó‡∫µ‡ªà‡ªÄ‡ªù‡∫≤‡∫∞‡∫™‡∫ª‡∫°
3. Observe - ‡∫™‡∫±‡∫á‡ªÄ‡∫Å‡∫î‡∫ú‡∫ª‡∫ô‡∫•‡∫±‡∫ö‡∫à‡∫≤‡∫Å‡ªÄ‡∫Ñ‡∫∑‡ªà‡∫≠‡∫á‡∫°‡∫∑
4. ‡∫ó‡∫≥‡∫ä‡ªâ‡∫≥‡∫à‡∫ª‡∫ô‡∫Å‡∫ß‡ªà‡∫≤‡∫à‡∫∞‡ªÑ‡∫î‡ªâ‡∫Ñ‡∫≥‡∫ï‡∫≠‡∫ö‡∫ó‡∫µ‡ªà‡∫™‡∫ª‡∫°‡∫ö‡∫π‡∫ô

‡∫Ñ‡∫≥‡∫ï‡∫≠‡∫ö‡∫™‡∫∏‡∫î‡∫ó‡ªâ‡∫≤‡∫ç‡ªÉ‡∫´‡ªâ‡ªÄ‡∫õ‡∫±‡∫ô‡∫û‡∫≤‡∫™‡∫≤‡∫•‡∫≤‡∫ß, ‡ªÉ‡∫ä‡ªâ Markdown ‡ªÉ‡∫´‡ªâ‡∫™‡∫ß‡∫ç‡∫á‡∫≤‡∫° ‡ªÅ‡∫•‡∫∞ ‡ªÉ‡∫´‡ªâ‡∫Ñ‡∫≥‡∫ï‡∫≠‡∫ö‡∫ó‡∫µ‡ªà‡ªÄ‡∫õ‡∫±‡∫ô‡∫õ‡∫∞‡ªÇ‡∫´‡∫ç‡∫î‡ªÅ‡∫•‡∫∞‡∫ä‡∫±‡∫î‡ªÄ‡∫à‡∫ô."""

    # ‡ªÄ‡∫û‡∫µ‡ªà‡∫° system message
    full_messages = [{"role": "system", "content": system_prompt}] + messages
    
    print(f"üí≠ ‡∫™‡∫ª‡ªà‡∫á‡∫Ç‡ªç‡ªâ‡∫Ñ‡∫ß‡∫≤‡∫°‡ªÑ‡∫õ‡∫´‡∫≤ LLM: {len(messages)} ‡∫Ç‡ªç‡ªâ‡∫Ñ‡∫ß‡∫≤‡∫°")
    
    # ‡ªÉ‡∫ä‡ªâ llm_with_tools ‡∫ó‡∫µ‡ªà‡∫°‡∫µ tools
    response = llm_with_tools.invoke(full_messages)
    
    print(f"ü§ñ ‡ªÑ‡∫î‡ªâ‡∫Æ‡∫±‡∫ö‡∫Å‡∫≤‡∫ô‡∫ï‡∫≠‡∫ö‡∫Å‡∫±‡∫ö‡∫à‡∫≤‡∫Å LLM")
    if hasattr(response, 'tool_calls') and response.tool_calls:
        print(f"üõ†Ô∏è LLM ‡∫ï‡ªâ‡∫≠‡∫á‡∫Å‡∫≤‡∫ô‡ªÉ‡∫ä‡ªâ‡ªÄ‡∫Ñ‡∫∑‡ªà‡∫≠‡∫á‡∫°‡∫∑: {[tc['name'] for tc in response.tool_calls]}")
    
    return {"messages": [response]}

def call_tools(state: AgentState):
    """‡ªÄ‡∫≠‡∫µ‡ªâ‡∫ô‡ªÉ‡∫ä‡ªâ‡ªÄ‡∫Ñ‡∫∑‡ªà‡∫≠‡∫á‡∫°‡∫∑‡∫ï‡∫≤‡∫°‡∫ó‡∫µ‡ªà LLM ‡∫Æ‡ªâ‡∫≠‡∫á‡∫Ç‡ªç"""
    messages = state["messages"]
    last_message = messages[-1]
    
    # ‡∫Å‡∫ß‡∫î‡∫™‡∫≠‡∫ö‡∫ß‡ªà‡∫≤‡∫°‡∫µ tool_calls ‡∫´‡∫º‡∫∑‡∫ö‡ªç‡ªà
    if hasattr(last_message, 'tool_calls') and last_message.tool_calls:
        print(f"üîß ‡∫Å‡∫≥‡∫•‡∫±‡∫á‡ªÄ‡∫≠‡∫µ‡ªâ‡∫ô‡ªÄ‡∫Ñ‡∫∑‡ªà‡∫≠‡∫á‡∫°‡∫∑: {[tc['name'] for tc in last_message.tool_calls]}")
        tool_results = execute_tools(last_message.tool_calls, tools)
        print(f"‚úÖ ‡ªÑ‡∫î‡ªâ‡∫ú‡∫ª‡∫ô‡∫•‡∫±‡∫ö‡∫à‡∫≤‡∫Å‡ªÄ‡∫Ñ‡∫∑‡ªà‡∫≠‡∫á‡∫°‡∫∑‡ªÅ‡∫•‡ªâ‡∫ß")
        return {"messages": tool_results}
    
    return {"messages": []}

# ===== Graph Construction =====
def create_flight_agent():
    """‡∫™‡ªâ‡∫≤‡∫á Flight Agent Graph ‡ªÅ‡∫ö‡∫ö‡∫á‡ªà‡∫≤‡∫ç"""
    workflow = StateGraph(AgentState)
    
    # ‡ªÄ‡∫û‡∫µ‡ªà‡∫° nodes
    workflow.add_node("agent", call_model)
    workflow.add_node("tools", call_tools)
    
    # ‡∫ï‡∫±‡ªâ‡∫á entry point
    workflow.set_entry_point("agent")
    
    # ‡ªÄ‡∫û‡∫µ‡ªà‡∫° conditional edges
    workflow.add_conditional_edges(
        "agent",
        should_continue,
        {
            "continue": "tools",
            "end": "__end__"  # ‡ªÉ‡∫ä‡ªâ string ‡ªÅ‡∫ó‡∫ô END constant
        }
    )
    
    # ‡ªÄ‡∫û‡∫µ‡ªà‡∫° edge ‡∫à‡∫≤‡∫Å tools ‡∫Å‡∫±‡∫ö‡∫Ñ‡∫∑‡∫ô agent
    workflow.add_edge("tools", "agent")
    
    # ‡ªÄ‡∫û‡∫µ‡ªà‡∫°‡∫Ñ‡∫ß‡∫≤‡∫°‡∫à‡∫≥
    memory = MemorySaver()
    
    return workflow.compile(checkpointer=memory)

class FlightBot:
    def __init__(self):
        self.app = create_flight_agent()
        self.thread_config = {"configurable": {"thread_id": "flight_assistant"}}
    
    def chat(self, message: str):
        """‡∫™‡∫ª‡∫ô‡∫ó‡∫∞‡∫ô‡∫≤‡∫Å‡∫±‡∫ö flight bot"""
        try:
            print(f"üõ©Ô∏è ‡∫Å‡∫≥‡∫•‡∫±‡∫á‡∫õ‡∫∞‡∫°‡∫ß‡∫ô‡∫ú‡∫ª‡∫ô: {message}\n")
            
            # ‡∫™‡ªâ‡∫≤‡∫á initial state
            initial_state = {
                "messages": [HumanMessage(content=message)],
                "next_action": ""
            }
            
            # ‡ªÅ‡∫•‡ªà‡∫ô agent
            result = self.app.invoke(initial_state, config=self.thread_config)
            
            # ‡ªÄ‡∫≠‡∫ª‡∫≤‡∫Ç‡ªç‡ªâ‡∫Ñ‡∫ß‡∫≤‡∫° AI ‡∫´‡∫•‡ªâ‡∫≤‡∫™‡∫∏‡∫î
            final_response = result["messages"][-1].content
            
            # ‡∫™‡∫∞‡ªÅ‡∫î‡∫á‡ªÉ‡∫ô‡∫Æ‡∫π‡∫ö‡ªÅ‡∫ö‡∫ö Markdown
            display(Markdown(f"## ‡∫Ñ‡∫≥‡∫ï‡∫≠‡∫ö‡∫à‡∫≤‡∫Å Flight Agent\n\n{final_response}"))
            
            return result
            
        except Exception as e:
            error_msg = f"‡ªÄ‡∫Å‡∫µ‡∫î‡∫Ç‡ªç‡ªâ‡∫ú‡∫¥‡∫î‡∫û‡∫≤‡∫î: {str(e)}"
            display(Markdown(f"**‡∫Ç‡ªç‡ªâ‡∫ú‡∫¥‡∫î‡∫û‡∫≤‡∫î:** {error_msg}"))
            return {"error": error_msg}
    
    def show_graph(self):
        """‡∫™‡∫∞‡ªÅ‡∫î‡∫á‡ªÇ‡∫Ñ‡∫á‡∫™‡ªâ‡∫≤‡∫á‡∫Å‡∫£‡∫≤‡∫ü"""
        try:
            # ‡∫™‡ªâ‡∫≤‡∫á‡ªÅ‡∫ú‡∫ô‡∫ú‡∫±‡∫á mermaid
            mermaid_png = self.app.get_graph().draw_mermaid_png()
            
            # ‡∫™‡∫∞‡ªÅ‡∫î‡∫á‡∫Æ‡∫π‡∫ö‡∫û‡∫≤‡∫ö
            display(Image(mermaid_png))
            
        except Exception as e:
            print(f"‡∫ö‡ªç‡ªà‡∫™‡∫≤‡∫°‡∫≤‡∫î‡∫™‡∫∞‡ªÅ‡∫î‡∫á‡∫Å‡∫£‡∫≤‡∫ü‡ªÑ‡∫î‡ªâ: {str(e)}")

# ‡ªÄ‡∫•‡∫µ‡ªâ‡∫°‡∫ï‡∫ª‡ªâ‡∫ô bot
def initialize_flight_bot():
    """‡ªÄ‡∫•‡∫µ‡ªâ‡∫°‡∫ï‡∫ª‡ªâ‡∫ô ‡ªÅ‡∫•‡∫∞ ‡∫™‡∫ª‡ªà‡∫á‡∫Å‡∫±‡∫ö‡∫Ñ‡∫∑‡∫ô flight bot"""
    bot = FlightBot()
    
    print("üõ©Ô∏è Flight Information Agent ‡∫û‡ªâ‡∫≠‡∫°‡ªÉ‡∫ä‡ªâ‡∫á‡∫≤‡∫ô‡ªÅ‡∫•‡ªâ‡∫ß! (‡ªÉ‡∫ä‡ªâ Groq LLaMA)")
    print("\n‡∫ó‡ªà‡∫≤‡∫ô‡∫™‡∫≤‡∫°‡∫≤‡∫î:")
    print("1. ‡∫Ñ‡∫ª‡ªâ‡∫ô‡∫´‡∫≤‡∫ó‡ªà‡∫Ω‡∫ß‡∫ö‡∫¥‡∫ô‡∫ó‡∫≤‡∫á‡∫î‡∫Ω‡∫ß - ‡∫û‡∫¥‡∫°: '‡∫Ñ‡∫ª‡ªâ‡∫ô‡∫´‡∫≤‡∫ó‡ªà‡∫Ω‡∫ß‡∫ö‡∫¥‡∫ô‡∫à‡∫≤‡∫Å VTE ‡ªÑ‡∫õ BKK ‡∫ß‡∫±‡∫ô‡∫ó‡∫µ 2025-08-26'")
    print("2. ‡∫Ñ‡∫ª‡ªâ‡∫ô‡∫´‡∫≤‡∫ó‡ªà‡∫Ω‡∫ß‡∫ö‡∫¥‡∫ô‡ªÑ‡∫õ‡∫Å‡∫±‡∫ö - ‡∫û‡∫¥‡∫°: '‡∫Ñ‡∫ª‡ªâ‡∫ô‡∫´‡∫≤‡∫ó‡ªà‡∫Ω‡∫ß‡∫ö‡∫¥‡∫ô‡ªÑ‡∫õ‡∫Å‡∫±‡∫ö‡∫à‡∫≤‡∫Å VTE ‡ªÑ‡∫õ BKK ‡∫ß‡∫±‡∫ô‡∫ó‡∫µ 26 ‡∫Å‡∫±‡∫ö‡∫ß‡∫±‡∫ô‡∫ó‡∫µ 30'")
    print("3. ‡∫Å‡∫ß‡∫î‡∫™‡∫≠‡∫ö‡∫Ç‡ªç‡ªâ‡∫°‡∫π‡∫ô‡∫Å‡∫≤‡∫ô‡∫à‡∫≠‡∫á - ‡∫û‡∫¥‡∫°: '‡∫Å‡∫ß‡∫î‡∫™‡∫≠‡∫ö‡∫Ç‡ªç‡ªâ‡∫°‡∫π‡∫ô‡∫Å‡∫≤‡∫ô‡∫à‡∫≠‡∫á‡∫à‡∫≤‡∫Å‡∫Æ‡∫π‡∫ö‡∫û‡∫≤‡∫ö'")
    print("\n‡ªÄ‡∫û‡∫∑‡ªà‡∫≠‡ªÄ‡∫ö‡∫¥‡ªà‡∫á‡ªÇ‡∫Ñ‡∫á‡∫™‡ªâ‡∫≤‡∫á‡∫Ç‡∫≠‡∫á Agent ‡∫û‡∫¥‡∫°: bot.show_graph()")
    print("üìå API Keys ‡∫ó‡∫µ‡ªà‡∫ï‡ªâ‡∫≠‡∫á‡∫Å‡∫≤‡∫ô: GROQ_API_KEY, SERPAPI_API_KEY, OPENROUTER_API_KEY")
    
    return bot

# ‡∫ï‡∫ª‡∫ß‡∫¢‡ªà‡∫≤‡∫á‡∫Å‡∫≤‡∫ô‡ªÉ‡∫ä‡ªâ‡∫á‡∫≤‡∫ô
if __name__ == "__main__":
    # ‡ªÄ‡∫•‡∫µ‡ªà‡∫°‡∫ï‡∫ª‡ªâ‡∫ô bot
    bot = initialize_flight_bot()
    
    # ‡∫™‡∫∞‡ªÅ‡∫î‡∫á‡ªÇ‡∫Ñ‡∫á‡∫™‡ªâ‡∫≤‡∫á‡∫Å‡∫£‡∫≤‡∫ü
    print("\nüìä ‡∫™‡∫∞‡ªÅ‡∫î‡∫á‡ªÇ‡∫Ñ‡∫á‡∫™‡ªâ‡∫≤‡∫á Agent Graph:")
    bot.show_graph()
    
    # ‡∫ï‡∫ª‡∫ß‡∫¢‡ªà‡∫≤‡∫á‡∫Å‡∫≤‡∫ô‡ªÉ‡∫ä‡ªâ‡∫á‡∫≤‡∫ô
    print("\nüîç ‡∫ï‡∫ª‡∫ß‡∫¢‡ªà‡∫≤‡∫á‡∫Å‡∫≤‡∫ô‡ªÉ‡∫ä‡ªâ‡∫á‡∫≤‡∫ô:")
    
    # ‡∫ï‡∫ª‡∫ß‡∫¢‡ªà‡∫≤‡∫á 1: ‡∫Ñ‡∫ª‡ªâ‡∫ô‡∫´‡∫≤‡∫ó‡ªà‡∫Ω‡∫ß‡∫ö‡∫¥‡∫ô
    # print("\n1. ‡∫Ñ‡∫ª‡ªâ‡∫ô‡∫´‡∫≤‡∫ó‡ªà‡∫Ω‡∫ß‡∫ö‡∫¥‡∫ô:")
    # bot.chat("‡∫Ñ‡∫ª‡ªâ‡∫ô‡∫´‡∫≤‡∫ó‡ªà‡∫Ω‡∫ß‡∫ö‡∫¥‡∫ô‡∫à‡∫≤‡∫Å VTE ‡ªÑ‡∫õ BKK ‡∫ß‡∫±‡∫ô‡∫ó‡∫µ 2025-08-26")
    
    # ‡∫ï‡∫ª‡∫ß‡∫¢‡ªà‡∫≤‡∫á 2: ‡∫ß‡∫¥‡ªÄ‡∫Ñ‡∫≤‡∫∞‡∫Æ‡∫π‡∫ö‡∫û‡∫≤‡∫ö‡∫Å‡∫≤‡∫ô‡∫à‡∫≠‡∫á  
    print("\n2. ‡∫ß‡∫¥‡ªÄ‡∫Ñ‡∫≤‡∫∞‡∫Æ‡∫π‡∫ö‡∫û‡∫≤‡∫ö‡∫Å‡∫≤‡∫ô‡∫à‡∫≠‡∫á:")
    bot.chat("‡∫ä‡ªà‡∫ß‡∫ç‡∫Å‡∫ß‡∫î‡∫™‡∫≠‡∫ö‡∫Ç‡ªç‡ªâ‡∫°‡∫π‡∫ô‡∫Å‡∫≤‡∫ô‡∫à‡∫≠‡∫á‡∫à‡∫≤‡∫Å‡∫Æ‡∫π‡∫ö‡∫û‡∫≤‡∫ö‡∫ó‡∫µ‡ªà‡ªÄ‡∫Å‡∫±‡∫ö‡ªÑ‡∫ß‡ªâ")
```
